```r
data=read.table("fishdata.txt",header=TRUE)
```

A)

```r
plot(x=data$Weight, y=data$Diagonal, main="Scatterplot of the relationship between weight and diagonal length", xlab="Weigths(in grams)", ylab="Diagonail length(in cm)")
```

B) The sample correlation between the weight and diagonal length:

```r
cor(x = data$Weight,y = data$Diagonal)
```

C) The Sample Correlation comes around to be +0.89, which can be seen as a strong positive non-linear relationship between the Weight and diagonal length

```r
plot(x=data$Weight, y=data$Diagonal, main="Scatterplot of the relationship between weight and diagonal length", xlab="Weigths(in grams)", ylab="Diagonail length(in cm)")
reg.data = lm(formula = Diagonal ~ Weight,data=data)
abline(reg=reg.data)
```

D) Yes, the sample correlation is good representaion of the relationship between Weight and Diagonal length as both scatterplot and regression line show a positive non-linear correlation.
