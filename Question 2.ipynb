```r
food=read.table("Food.txt", header=TRUE)
```

```r
Pesticide=read.table("Pesticide.txt", header=TRUE)
```

```r
total=read.table("Total.txt", header=TRUE)
```

```r
data = data.frame(Food = food,Total=total,Pesticide = Pesticide)
```

```r
head(data)
```

A)

```r
barplot(height = data$Pesticide, names.arg = data$Food, main = "Bar Chart of fruit/vegetable that tested positive for pesticides",xlab="Fruits/vegetables",ylab="frequency")
```

B)

```r
data$Proportion=data$Pesticide/data$Total
```

```r
barplot(height = data$Proportion, names.arg = data$Food, main = "Bar Chart of fruit/vegetable that tested positive for pesticides",xlab="Fruits/vegetables",ylab="relative frequency")
```
